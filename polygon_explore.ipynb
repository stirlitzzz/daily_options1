```julia
using CSV
using DataFrames
using Dates
using TimeZones  # For time zone conversions

# Load the CSV file
file_path = "data/polygon_spy_w_headers.csv"  # Replace with your actual file path
data = CSV.read(file_path, DataFrame)

# Convert `window_start` from Unix time (in nanoseconds) to seconds
data.window_start = data.window_start ./ 1e9

# Convert to DateTime in UTC
#data.window_start = unix2datetime.(data.window_start)
data.window_start = DateTime.(unix2datetime.(data.window_start))

# Convert from UTC to Eastern Standard Time (EST)
est = TimeZone("America/New_York")
data.window_start = ZonedDateTime.(data.window_start, est)

# Pivot the data so each ticker becomes a column
pivot_data = unstack(data, :window_start, :ticker, :close)

# Sort the data by `window_start`
pivot_data = sort(pivot_data, :window_start)

# Display the result
println(pivot_data)
```
